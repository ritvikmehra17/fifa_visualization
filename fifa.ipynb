```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("input/fifa19_data.csv")
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset=pd.DataFrame(list(zip(dataset.Potential.head(20))),
                    columns=["Potentials"],
                    index=dataset.Name.head(20))
dataset.plot(kind="bar")
```

```python
r=dataset["Release Clause"]

plt.bar(r.tail(10),dataset.Name.tail(10))
plt.title("highest paid players")
plt.ylabel("Player Name")
plt.xlabel("Release Clause")
```

```python
plt.barh(dataset.Nationality.head(20), dataset.Name.head(20))
plt.ylabel("Nationality")
plt.xlabel("Player Name")
plt.figure(figsize=(15,3))
```

```python
plt.bar(dataset.Wage.head(20), dataset.Name.head(20))
plt.ylabel("Wage")
plt.xlabel("Player Name")
plt.figure(figsize=(15,3))
```

```python
plt.bar(dataset["Weak Foot"].head(20), dataset.Name.head(20))
plt.ylabel("Wage")
plt.xlabel("Player Name")
plt.figure(figsize=(15,3))
```

```python
plt.barh( dataset.Name.head(20),dataset["Skill Moves"].head(20))
plt.ylabel("Moves")
plt.xlabel("Player Name")
plt.title("Skills of Players")
```
